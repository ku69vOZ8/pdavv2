【例10-11】函数极值求解。

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import fmin,fminbound  
def f(x):    
     return x**2+10*np.sin(x)+1
x = np.linspace(-10,10,num = 500)
min1 = fmin(f,3)  #求3附近的极小值
min2 = fmin(f,0)  #求0附近的极小值
min_global = fminbound(f,-10,10) #这个区域的最小值
print(min1)
print(min2)
print(min_global)
plt.plot(x,f(x))
plt.show()
```

【例10-13】分别使用1阶、3阶和9阶多项式对数据进行拟合。

```python
x = np.linspace(0,np.pi*2)
y = np.sin(x)
f1 = np.poly1d(np.polyfit(x,y,1))
f3 = np.poly1d(np.polyfit(x,y,3))
f9 = np.poly1d(np.polyfit(x,y,9))
t = np.linspace(-3*np.pi,np.pi*3,80)
plt.plot(x,y,'x',t,f1(t),':',t,f3(t),'--',t,f9(t),'-.')
plt.axis([0,3*np.pi,-1.5,1.5])
plt.legend(['data',r'$n = 1$',r'$n = 3$',r'$n = 9$'])
plt.show()
```

【例10-14】最小二乘估计示例。

```python
from scipy import optimize
def myfunc(x,a,b,w,t):
    return a*np.exp(-b*np.sin(w*x+t))
x = np.linspace(0,2*np.pi)
par = [3,2,1.25,np.pi/4]
y = myfunc(x,*par)
y_noise = y+0.8*np.random.randn(len(y))
def errf(p,x,y):
    return y-myfunc(x,*p)
c,rv = optimize.leastsq(errf,[1,1,1,1],args = (x,y_noise))
#c返回找到的最小二乘估计
plt.plot(x,y_noise,'x',x,y,x,myfunc(x,*c),':')
plt.legend(['data','actual','leastsq'])
plt.show()
```

【例10-15】曲线拟合示例。

```python
p,e = optimize.curve_fit(myfunc,x,y_noise)
print('p是对参数的估计值:\n',p)
print('e是4个估计参数的协方差矩阵:\n',e)
```

## 本章实训

利用SVD分解实现图像数据的压缩

压缩原理（SVD的降维原理）





奇异值可以被看作成一个矩阵的代表值，或者说，奇异值能够代表这个矩阵的信息。

当奇异值越大时，它代表的信息越多。因此，我们取前面若干个最大的奇异值，就可以基本上还原出数据本身。

# 打开图像

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
im = np.array(Image.open('data/lena_gray.tif'))
#打开图像数据并转换为数组
plt.imshow(im,cmap = 'Greys_r')
plt.title("src")
plt.axis('off')
```

# 对图像矩阵SVD分解

```python
U,sigma,VT = np.linalg.svd(im)
print(U)
print('前30个特征值为：\n',sigma[:30])
```

#分别只利用30个特征值和120个特征值重构图像

```python
c = 30
r1 = (U[:,:c]).dot(np.diag(sigma[:c])).dot(VT[:c,:])
c = 120
r2 = (U[:,:c]).dot(np.diag(sigma[:c])).dot(VT[:c,:])
plt.subplot(1,2,1)
plt.imshow(r1.astype(np.uint8),cmap = 'Greys_r')
plt.title("nums_sigma = 30")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(r2.astype(np.uint8),cmap = 'Greys_r')
plt.title("nums_sigma = 120")
plt.axis('off')
```
